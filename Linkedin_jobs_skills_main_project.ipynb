<h3 style="font-weight: bold;color:#07306c"> 2. Data Pre-processing</h3>

 <h4 style="font-weight: bold;color:red">1. Handling missing values </h4>

<h5 style="font-weight: bold;color:#006dd9">=> Fill <span style="color:green">Company</span> with string <span style="color:green">Confidential</span></h5> 

```python
df_merged_job_post['company'] = df_merged_job_post['company'].fillna('Confidential')
```

```python
print(df_merged_job_post.isnull().sum())
```

<h5 style="font-weight: bold;color:#006dd9"> => <span style="color:green">Drop</span> rows with null values in the <span style="color:green">job_location</span> column</h5> 

```python
df_merged_job_post.dropna(subset=['job_location'], inplace=True)
# Check null after drop rows from job location
print(f"Null values after inplace: {df_merged_job_post['job_location'].isnull().sum()}")
```

<h5 style="font-weight: bold;color:#006dd9">   => Fill <span style="color:green">Job_Summary</span> with <span style="color:green">N/A</span></h5>

```python
df_merged_job_post['job_summary'] = df_merged_job_post['job_summary'].fillna('N/A')
```

<h5 style="font-weight: bold;color:#006dd9">=> Fill <span style="color:green">skills_list</span> with empty list <span style="color:green">[]</span></h5>

```python
#df_merged_job_post['skills_list'] = df_merged_job_post['skills_list'].apply(lambda x: x if isinstance(x, list) else [])
df_merged_job_post['skills_list'] = df_merged_job_post['skills_list'].apply(
    lambda x: x if isinstance(x, list) else ([] if pd.isna(x) else x)
)
```

```python
print(df_merged_job_post.isnull().sum())
```

 <h4 style="font-weight: bold;color:red">2. Removing duplicates </h4>

```python
# Shape of DataFrame After deleting null rows from job location
df_merged_job_post.shape
```

```python
# Check before removal
print("BEFORE:")
print(f"Total rows: {len(df_merged_job_post)}")
duplicate_count = len(df_merged_job_post) - df_merged_job_post['job_link'].nunique()
print(f"Duplicate job links: {duplicate_count}")
# Remove duplicates
df_merged_job_post = df_merged_job_post.drop_duplicates(subset=['job_link'], keep='first')

# Check after removal
print("\nAFTER:")
print(f"Total rows: {len(df_merged_job_post)}")
print(f"Rows removed: {duplicate_count}")
```

 <h4 style="font-weight: bold;color:red">3. Convert to proper case </h4>

```python
# List of columns to convert to proper case
text_columns = [
 'job_title',
 'company',
 'job_location',
 'search_city',
 'search_country',
 'search_position',
 'job_level',
 'job_type',
]

for col in text_columns:
    if col in df_merged_job_post.columns:
        df_merged_job_post[col] = df_merged_job_post[col].str.title()
        print(f"Converted {col} to proper case")
```

 <h4 style="font-weight: bold;color:red">3. Correcting data types </h4>

<h5 style="font-weight: bold;color:#006dd9">=> Convert to "String"</h5>

```python
# List of columns to convert to string
string_columns = ['job_link',
 'job_title',
 'company',
 'job_location',
 'search_city',
 'search_country',
 'search_position',
 'job_level',
 'job_type',
 'job_summary'
 ]

for col in string_columns:
    if col in df_merged_job_post.columns:
        df_merged_job_post[col] = df_merged_job_post[col].astype('string')
        print(f"Converted {col} to string dtype")
```

<h5 style="font-weight: bold;color:#006dd9">=> Convert to "Boolean"</h5>

```python
# Convert 't'/'f'  to boolean
df_merged_job_post['got_summary'] = df_merged_job_post['got_summary'].str.upper().map({'T': True, 'F': False})
df_merged_job_post['got_ner'] = df_merged_job_post['got_ner'].str.upper().map({'T': True, 'F': False})
df_merged_job_post['is_being_worked'] = df_merged_job_post['is_being_worked'].str.upper().map({'T': True, 'F': False})

df_merged_job_post.head(5)
```

```python
df_merged_job_post['last_processed_time'] = pd.to_datetime(df_merged_job_post['last_processed_time'])
print(df_merged_job_post['last_processed_time'].dtype)
```

<h5 style="font-weight: bold;color:#006dd9">=> Convert first_seen object to "date For DD/MM/YYYY format"</h5>

```python
df_merged_job_post['first_seen'] = pd.to_datetime(df_merged_job_post['first_seen'], format='%d/%m/%Y')
print(df_merged_job_post['first_seen'].dtype)
```

<span style="font-weight: bold;color:#2113a8;font-size:20px"><u>Summary of Findings</u><span>

<h3>🔑 Skills Landscape</h3>
<ul>
<li>
<p><strong>Soft skills dominate</strong>: Communication, Teamwork, Leadership, and Customer Service are the top-requested skills.</p>
</li>
<li>
<p><strong>Technical skills</strong> like Microsoft Office, Project Management, and Data Analysis also appear but at lower frequencies.</p>
</li>
<li>
<p>Skill requirements are <strong>uniform across job levels</strong> &ndash; both Associate and Mid-Senior roles average ~400 skills per posting, which is unexpectedly high.</p>
</li>
</ul>
<hr />
<h3>🌍 Geographic Distribution</h3>
<ul>
<li>
<p><strong>United States dominates</strong>: ~85% of postings and searches, especially in New York, Houston, Chicago, and Los Angeles.</p>
</li>
<li>
<p><strong>Other top countries</strong>: UK, Canada, India, and Australia, but at much smaller scales.</p>
</li>
<li>
<p><strong>Remote jobs</strong> are almost negligible compared to onsite (~1.2M onsite vs ~3K remote), suggesting a market still tilted towards physical presence.</p>
</li>
</ul>
<hr />
<h3>👔 Role-Based Insights</h3>
<ul>
<li>
<p><strong>Mid-Senior roles dominate</strong>: ~900,000 postings vs ~120,000 Associate-level jobs.</p>
</li>
<li>
<p>High-demand roles like <strong>Account Executive, Data Scientist, Software Engineer, and Healthcare professionals</strong> appear frequently.</p>
</li>
<li>
<p>Companies prefer candidates with <strong>experience who can deliver immediately</strong>, reducing focus on entry-level hiring.</p>
</li>
</ul>
<hr />
<h3>⏳ Temporal Trends</h3>
<ul>
<li>
<p>Job postings peak around <strong>Jan 13&ndash;14, 2024</strong> (~400,000 in a single day), then drop sharply.</p>
</li>
<li>
<p>This indicates <strong>posting bursts or campaigns</strong> rather than steady daily hiring.</p>
</li>
<li>
<p>Onsite jobs follow this trend strongly, while Remote/Hybrid remain flat.</p>
</li>
</ul>
<hr />
<h3>🏢 Company Analysis</h3>
<ul>
<li>
<p><strong>Health Ecareers</strong> leads with 30,000+ postings, followed by <strong>Jobs for Humanity</strong> (~21,000).</p>
</li>
<li>
<p>Other active companies: Travelnursesource, Dollar General, McDonald&rsquo;s, Jobot.</p>
</li>
<li>
<p><strong>Variation in skill lists</strong>: Some companies post consistent, simple requirements; others list hundreds of skills (noise/outliers).</p>
</li>
</ul>
<hr />
<h3>📌 Overall Findings for the Aim</h3>
<ol>
<li>
<p>The <strong>2024 job market is highly skewed</strong> towards onsite, mid-senior roles, particularly in the U.S.</p>
</li>
<li>
<p><strong>Soft skills remain universally in demand</strong>, cutting across industries and job levels.</p>
</li>
<li>
<p><strong>Remote work opportunities are still rare</strong>, contradicting expectations of post-pandemic flexibility.</p>
</li>
<li>
<p>Hiring is concentrated among <strong>a few big players</strong>, making the landscape uneven.</p>
</li>
<li>
<p><strong>Posting activity is burst-driven</strong>, suggesting companies push out jobs in large batches.</p>
</li>
</ol>

<span style="font-weight: bold;color:#2113a8;font-size:20px"><u>Recommendation</u><span>

<ul>
<li>
<p><strong>Standardize Skills Taxonomy</strong></p>
<ul>
<li>
<p>Consolidate duplicate/overlapping skill terms (e.g., &ldquo;MS Office&rdquo; vs &ldquo;Microsoft Office&rdquo;).</p>
</li>
<li>
<p>Helps reduce noise and improves clarity for both job seekers and employers.</p>
</li>
</ul>
</li>
<li>
<p><strong>Encourage Structured Job Postings</strong></p>
<ul>
<li>
<p>Use templates for roles, levels, and required skills.</p>
</li>
<li>
<p>Makes postings more consistent and easier to analyze.</p>
</li>
</ul>
</li>
<li>
<p><strong>Expand Remote/Hybrid Opportunities</strong></p>
<ul>
<li>
<p>Companies could attract wider talent pools by offering flexible options.</p>
</li>
<li>
<p>Particularly useful in IT, data, and creative roles.</p>
</li>
</ul>
</li>
<li>
<p><strong>Support Entry-Level Hiring</strong></p>
<ul>
<li>
<p>Balance the current mid-senior focus with more Associate/Graduate roles.</p>
</li>
<li>
<p>Encourages skill development and sustains long-term workforce pipelines.</p>
</li>
</ul>
</li>
<li>
<p><strong>Enhance Data Quality on Platforms</strong></p>
<ul>
<li>
<p>Deduplicate postings, validate job details, and improve metadata.</p>
</li>
<li>
<p>Ensures reliable insights for researchers, job seekers, and recruiters.</p>
</li>
</ul>
</li>
</ul>
